# Análise de Idade Materna e Peso ao Nascimento dos Bebês em Rondônia - 2019

## Introdução

Este relatório apresenta uma análise dos dados de nascidos vivos em Rondônia no ano de 2019, com foco nas variáveis **idade da mãe** e **peso ao nascimento dos bebês**. O objetivo é identificar padrões que possam orientar **políticas públicas** e melhorar a saúde materno-infantil no estado. A base de dados utilizada é a **SINASC** fornecida pelo **DataSUS**, fonte oficial de dados de nascidos vivos no Brasil.

A análise busca destacar pontos relevantes que possam embasar a criação de ações estratégicas na área da saúde, ajudando a compreender melhor os fatores que influenciam a saúde dos recém-nascidos e das mães.

## 1. Idade Materna

### Estatísticas Descritivas

- **Média**: 26,09 anos
- **Mediana**: 26 anos
- **Desvio Padrão**: 6,39 anos
- **Idade mínima**: 11 anos
- **Idade máxima**: 53 anos
- **Faixa Interquartil** (25%-75%): 21 a 31 anos

### Análise

A idade média das mães é de 26 anos, o que está dentro da faixa etária considerada ideal para a gestação. No entanto, observa-se que há mães com apenas 11 anos, o que aponta para a necessidade de maior atenção em relação às **gravidezes precoces**. Essas gravidezes podem trazer complicações para a saúde da mãe e do bebê, além de estar relacionadas a questões socioeconômicas e de falta de educação sexual adequada.

### Recomendação para Políticas Públicas

Dada a presença de casos de gravidez em idades tão jovens, recomenda-se o fortalecimento de políticas voltadas à **educação sexual** nas escolas e à **disponibilização de métodos contraceptivos** de forma acessível. Programas de **planejamento familiar** também podem ajudar a reduzir a incidência de gravidezes não planejadas, especialmente entre adolescentes.

## 2. Peso ao Nascimento dos Bebês

### Estatísticas Descritivas

- **Média**: 3243 gramas
- **Mediana**: 3270 gramas
- **Desvio Padrão**: 544,75 gramas
- **Peso mínimo**: 258 gramas
- **Peso máximo**: 5985 gramas
- **Faixa Interquartil** (25%-75%): 2965 a 3575 gramas

### Análise

O peso médio ao nascer é de 3243 gramas, o que se enquadra dentro da faixa saudável para recém-nascidos. No entanto, o peso mínimo de 258 gramas aponta para casos graves, possivelmente de **nascimentos prematuros** ou com complicações severas. O acompanhamento adequado durante o pré-natal pode ajudar a identificar esses casos de risco e atuar preventivamente para melhorar as chances de um desfecho saudável.

### Recomendação para Políticas Públicas

Para garantir que os bebês nasçam com peso adequado, é crucial reforçar o acesso a **cuidados pré-natais de qualidade** e **programas de nutrição materna**. A promoção de campanhas educativas para mães, sobre a importância de um acompanhamento contínuo durante a gestação, pode reduzir a incidência de baixo peso ao nascer. Além disso, políticas que ampliem o número de consultas pré-natais e garantam atendimento de qualidade, especialmente em regiões mais vulneráveis, são essenciais.

## 3. Conclusão

A análise dos dados de nascidos vivos em Rondônia revela que, apesar de a maioria dos nascimentos ocorrer dentro de parâmetros saudáveis, há desafios importantes a serem enfrentados, como as **gravidezes precoces** e os **casos de baixo peso ao nascer**. Esses desafios exigem políticas públicas focadas em **educação sexual**, **planejamento familiar** e **cuidados pré-natais**.

Investir em **educação** e **saúde preventiva** pode contribuir significativamente para a redução de complicações associadas à gestação e ao nascimento, resultando em melhor qualidade de vida para mães e bebês.
