#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Importar as bibliotecas necessárias
import pandas as pd               # Para manipulação de dados
import seaborn as sns             # Para visualização de dados
import matplotlib.pyplot as plt    # Para criação de gráficos

# Carregar a base de dados SINASC
# Altere o caminho para o local onde você armazenou o arquivo CSV
sinasc_df = pd.read_csv('SINASC_RO_2019.csv')

# Limpar os nomes das colunas, removendo espaços em branco
sinasc_df.columns = sinasc_df.columns.str.strip()

# Calcular a média do APGAR5 para cada categoria de CONSULTAS
apgar5_medio = sinasc_df.groupby('CONSULTAS')['APGAR5'].mean().reset_index()

# Criar um gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x='CONSULTAS', y='APGAR5', data=apgar5_medio, palette='viridis')

# Adicionar título e rótulos aos eixos
plt.title('APGAR5 Médio por Número de Consultas', fontsize=16)
plt.xlabel('Número de Consultas', fontsize=14)
plt.ylabel('APGAR5 Médio', fontsize=14)

# Exibir o gráfico
plt.xticks(rotation=45)  # Rotacionar os rótulos do eixo x, se necessário
plt.tight_layout()        # Ajustar layout para melhor apresentação
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Importar as bibliotecas necessárias
import pandas as pd               # Para manipulação de dados
import seaborn as sns             # Para visualização de dados
import matplotlib.pyplot as plt    # Para criação de gráficos

# Carregar a base de dados SINASC
# Altere o caminho para o local onde você armazenou o arquivo CSV
sinasc_df = pd.read_csv('SINASC_RO_2019.csv')

# Limpar os nomes das colunas, removendo espaços em branco
sinasc_df.columns = sinasc_df.columns.str.strip()

# Criar um painel de gráficos (2 linhas x 2 colunas)
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Gráfico de barras para SEXO
sns.countplot(x='SEXO', data=sinasc_df, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Distribuição de SEXO', fontsize=16)
axes[0, 0].set_ylabel('Frequência', fontsize=12)

# Gráfico de barras para GESTACAO
sns.countplot(x='GESTACAO', data=sinasc_df, ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title('Distribuição de GESTACAO', fontsize=16)
axes[0, 1].set_ylabel('Frequência', fontsize=12)

# Gráfico de barras para ESTCIVMAE
sns.countplot(x='ESTCIVMAE', data=sinasc_df, ax=axes[1, 0], palette='viridis')
axes[1, 0].set_title('Distribuição de Estado Civil da Mãe', fontsize=16)
axes[1, 0].set_ylabel('Frequência', fontsize=12)

# Gráfico de barras para CONSULTAS
sns.countplot(x='CONSULTAS', data=sinasc_df, ax=axes[1, 1], palette='viridis')
axes[1, 1].set_title('Distribuição de CONSULTAS', fontsize=16)
axes[1, 1].set_ylabel('Frequência', fontsize=12)

# Ajustar layout para melhor apresentação
plt.tight_layout()

# Exibir o painel de gráficos
plt.show()
```

# Relatório sobre a Análise de Dados de Nascimentos - SINASC RO 2019

## Introdução

Este relatório apresenta uma análise dos dados de nascimentos no estado de Rondônia em 2019, com foco nas variáveis relacionadas ao estado civil da mãe, sexo do bebê, gestação e número de consultas pré-natais. A análise foi realizada utilizando a base de dados do Sistema de Informações sobre Nascidos Vivos (SINASC), permitindo uma compreensão mais aprofundada das características dos nascimentos na região.

Diante do cenário atual, é crucial enfatizar a importância de políticas públicas que promovam a saúde infantil e o bem-estar das crianças, uma vez que o cuidado adequado durante a gestação e o início da vida é fundamental para o desenvolvimento saudável.

## 1. Contagem de Estado Civil da Mãe

A análise da variável **Estado Civil da Mãe (ESTCIVMAE)** revela as seguintes frequências:

| Estado Civil       | Frequência |
|--------------------|------------|
| 2.0 (Casada)       | 9924       |
| 1.0 (Solteira)     | 8952       |
| 5.0 (Divorciada)   | 7180       |
| 4.0 (Viúva)        | 453        |
| 9.0 (Ignorado)     | 145        |
| 3.0 (Separada)     | 57         |

**Interpretação**: A maior parte dos nascimentos ocorreu entre mães casadas (aproximadamente 41% do total), seguidas por mães solteiras (cerca de 36%). O número significativo de mães divorciadas indica uma mudança nas dinâmicas familiares, ressaltando a necessidade de suporte e políticas que atendam às diversas situações familiares.

## 2. Distribuição de Sexo do Bebê

A análise da variável **Sexo** apresenta as seguintes frequências:

| Sexo       | Frequência |
|------------|------------|
| Masculino  | 13886      |
| Feminino   | 13138      |

**Interpretação**: Há uma ligeira predominância de nascimentos masculinos em relação aos femininos, com uma diferença de aproximadamente 748 bebês, o que é consistente com padrões demográficos típicos.

## 3. Distribuição de Gestações

A variável **Gestação** foi analisada e os resultados foram:

| Gestações             | Frequência |
|-----------------------|------------|
| 37 a 41 semanas       | 22589      |
| 32 a 36 semanas       | 2249       |
| 42 semanas e mais     | 700        |
| 28 a 31 semanas       | 165        |
| 22 a 27 semanas       | 86         |
| Menos de 22 semanas    | 7          |

**Interpretação**: A grande maioria das gestações (cerca de 90%) ocorreu entre 37 a 41 semanas, o que é considerado o intervalo ideal para o nascimento. A presença de gestações prematuras, embora em números pequenos, destaca a necessidade de intervenções e programas de saúde pública que garantam cuidados adequados para gestantes.

## 4. Número de Consultas Pré-Natais

A análise da variável **Consultas** indica a seguinte distribuição:

| Consultas | Frequência |
|-----------|------------|
| 4         | 18624      |
| 3         | 5988       |
| 2         | 1379       |
| 1         | 986        |
| 9 (Ignorado) | 51      |

**Interpretação**: A maioria das mães teve 4 consultas pré-natais, o que é um indicativo positivo de cuidados adequados durante a gravidez. No entanto, a presença de mães com apenas uma consulta destaca a necessidade de programas de conscientização e acesso a cuidados pré-natais, fundamentais para garantir a saúde da criança e da mãe.

## 5. Média do APGAR5 por Número de Consultas

Os resultados da média do **APGAR5** por número de consultas pré-natais são:

| Consultas | APGAR5 Médio |
|-----------|--------------|
| 1         | 8.98         |
| 2         | 9.06         |
| 3         | 9.15         |
| 4         | 9.21         |
| 9         | 8.86         |

**Interpretação**: Observa-se que o APGAR5 médio aumenta com o número de consultas pré-natais, indicando que um acompanhamento pré-natal mais intenso está associado a melhores resultados de saúde para os recém-nascidos. Isso reforça a importância do pré-natal adequado e da implementação de políticas que incentivem o acompanhamento médico regular.

## Conclusão

A análise dos dados do SINASC de Rondônia em 2019 fornece insights valiosos sobre as características demográficas e de saúde das mães e dos recém-nascidos. As informações revelam tendências sobre o estado civil, sexo, gestação e a importância das consultas pré-natais, com implicações significativas para políticas de saúde pública.

É imperativo que as autoridades de saúde considerem estas evidências ao desenvolver programas que visem aumentar o acesso e a conscientização sobre a importância do pré-natal e dos cuidados com a saúde das crianças. Investir na saúde infantil não é apenas uma responsabilidade social, mas um passo fundamental para garantir um futuro mais saudável e promissor para a sociedade como um todo.
