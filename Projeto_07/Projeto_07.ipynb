#### Marcação de bom e mau
O objetivo da modelagem é classificar o risco de inadimplência, ou como se diz no meio, o risco de *default*. Podemos fazer longas discussões sobre o conceito de *default* com base em estudos e exigências regulatórias, para efeitos deste estudo, um cliente em *default* é aquele que está em 60 dias de atraso ou mais. Então classificaremos os clientes como 'bons' e 'maus' assim:
- **Maus** pagadores: são aqueles que entraram em 'default' (atraso 60 dias ou mais) nos 24 meses seguintes à aquisição do cartão de crédito.
- **Bons** pagadores: são considerados todos os demais.
- **Excluídos**: Clientes que não adquiriram um cartão de crédito (seja por recusa, seja por desistência) não possuem informações de pagamento, portanto não se pode identificar se são bons ou maus. Há uma longa discussão e literatura sobre *inferência de rejeitados* que está fora do escopo deste exercício.

#### Construindo a variável resposta
A base de pagamentos está em um formato de 'base larga'. Essa base possui informações de pagamentos do cliente mês a mês a partir do mês de aquisição do crédito (mês 0) até o vigésimo quarto mês após a aquisição do crédito (mês 24). Utilizaremos essa base para determinar se um proponente é considerado 'bom pagador' ou caso apresente atraso representativo, será considerado 'mau pagador'.

#### Base larga vs base longa
A base ser larga significa que há uma linha para cada cliente, e que as informações estarão nas colunas, em contraste com a 'base longa', em que haveria uma linha para cada combinação cliente/mês, uma coluna indicando o cliente, outra indicando o mês, e apenas uma coluna com a informação do atraso.

````python
import pandas as pd
````

````python
propostas = pd.read_csv('application_record.csv')
pg = pd.read_csv('pagamentos_largo.csv')
````

````python
pg.head()
````

````python
# 1) Marcar default no mês.

# Carregar a base de pagamentos
pagamentos = pd.read_csv('pagamentos_largo.csv')

# Criar a coluna default que indica se o cliente está em default em cada mês
# Atraso de 60 dias ou mais é representado pelos valores 2, 3, 4 ou 5
default_indicador = pagamentos.iloc[:, 1:25].isin([2, 3, 4, 5]).astype(int)

# Adicionando a coluna de default a cada mês
pagamentos = pd.concat([pagamentos, default_indicador.add_suffix('_default')], axis=1)

# Exibir os primeiros registros para verificação
print("Tarefa 1: Indicador de default por mês")
print(pagamentos.head())
````

````python
# 2) Bons' e 'Maus' pagadores ao longo de todos os 24 meses de desempenho.

pagamentos['tem_default'] = default_indicador.sum(axis=1) > 0

# Print da contagem de defaults na base de pagamentos

print("\nContagem de clientes por default na base de pagamentos:")
print(pagamentos['tem_default'].value_counts())
````

````python
# 3) Marcando proponentes expostos ao risco de crédito.

clientes_validos = propostas[propostas['ID'].isin(pagamentos['ID'])]

# Adicionar a informação de default à base de propostas
resultado_final = pd.merge(clientes_validos, pagamentos[['ID', 'tem_default']], on='ID', how='left')

# Verificar se houve algum problema com os dados
resultado_final['tem_default'].fillna(False, inplace=True)  # Preencher valores NaN com False

# Tarefa 2: Marcando bons e maus pagadores
resultado_final['mau_pagador'] = resultado_final['tem_default']  # Usando a coluna tem_default para definir maus pagadores

# Contar os valores de 'mau_pagador'
contagem_default = resultado_final['mau_pagador'].value_counts()

# Print dos resultados finais
print("\nTarefa 2: Contagem de bons e maus pagadores")
print(contagem_default)

````

````python
# 4) Consolidando as informações.

resultado_final = propostas.merge(pagamentos[['ID', 'tem_default']], on='ID', how='inner')

# Preencher valores NaN com False na coluna 'tem_default'
resultado_final['tem_default'].fillna(False, inplace=True)

# Print dos resultados da junção (Tarefa 4)
print("\nTarefa 4: Informações consolidadas após junção")
print(resultado_final.head())  # Visualizar os primeiros registros
````

````python
# 5)  Contagem dos valores do default.
contagem_default = resultado_final['tem_default'].value_counts()

# Adicionar a coluna de 'mau_pagador'
resultado_final['mau_pagador'] = resultado_final['tem_default']

# Print da contagem de bons e maus pagadores (Tarefa 5)
print("\nTarefa 5: Contagem de bons e maus pagadores")
print(contagem_default)
````

Conclusão sobre Bons e Maus Pagadores no Risco de Crédito

Ao analisar o risco de inadimplência em um projeto de classificação de crédito, identificamos os principais fatores que definem bons e maus pagadores:
Maus Pagadores

    Inadimplência: Um cliente é considerado um mau pagador quando apresenta atraso de 60 dias ou mais no pagamento (default) em até 24 meses após a aquisição do crédito.
    Atrasos repetidos: Aqueles que acumulam vários meses de atraso significativo são mais propensos a inadimplência crônica.

Bons Pagadores

    Pagamentos pontuais: Clientes que mantêm seus pagamentos em dia ao longo dos 24 meses, sem atrasos maiores que 30 dias, são considerados bons pagadores.
    Perfil financeiro estável: Indicadores como renda estável, baixo endividamento e tempo de emprego mais longo estão correlacionados com um comportamento de pagamento mais confiável.

Fatores Relevantes

    Perfil econômico: Clientes com maior renda e estabilidade no emprego tendem a ser bons pagadores.
    Características familiares: Solicitantes com mais filhos ou maiores responsabilidades financeiras podem enfrentar maior risco de inadimplência.
    Histórico de crédito: Clientes que já possuem outros empréstimos ou cartões podem apresentar maior risco dependendo do comportamento passado.

A análise de dados detalhada desses fatores é essencial para classificar corretamente o risco e garantir a precisão dos modelos de crédito.
