1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

# Carregando a base de dados
df = pd.read_csv('previsao_de_renda.csv')

# Convertendo a coluna data_ref para datetime
df['data_ref'] = pd.to_datetime(df['data_ref'])

# Criando a coluna 'ano'
df['ano'] = df['data_ref'].dt.year

# Lista de variáveis qualitativas
variaveis_qualitativas = ['sexo', 'posse_de_veiculo', 'posse_de_imovel',
                          'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']

# Criando gráficos para cada variável qualitativa
for var in variaveis_qualitativas:
    # Agrupando os dados por ano e pela variável qualitativa
    dist_ano = df.groupby(['ano', var]).size().unstack()

    # Plotando o gráfico de barras para a variável atual
    plt.figure(figsize=(10, 5))
    dist_ano.plot(kind='bar', stacked=True, colormap='Set2')
    plt.title(f'Distribuição de {var} ao Longo dos Anos')
    plt.ylabel('Número de Pessoas')
    plt.xlabel('Ano')
    plt.legend(title=var, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # Imprimindo a tabela de distribuição ao final
    print(f"\nDistribuição de {var} por ano:")
    print(dist_ano)
````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

# Carregando a base de dados
df = pd.read_csv('previsao_de_renda.csv')

# Convertendo a coluna data_ref para datetime
df['data_ref'] = pd.to_datetime(df['data_ref'])

# Criando a coluna 'ano'
df['ano'] = df['data_ref'].dt.year

# Lista de variáveis qualitativas
variaveis_qualitativas = ['sexo', 'posse_de_veiculo', 'posse_de_imovel',
                          'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']

# Loop para criar gráficos de médias de renda para cada nível das variáveis qualitativas
for var in variaveis_qualitativas:
    # Agrupando os dados por ano e pela variável qualitativa e calculando a média de renda
    media_renda = df.groupby(['ano', var])['renda'].mean().unstack()

    # Plotando o gráfico de linha para cada variável qualitativa
    plt.figure(figsize=(10, 5))
    media_renda.plot(kind='line', marker='o')
    plt.title(f'Média de Renda por {var} ao Longo dos Anos')
    plt.ylabel('Média de Renda')
    plt.xlabel('Ano')
    plt.legend(title=var, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # Imprimindo os dados finais de média de renda por ano e por variável qualitativa
    print(f"\nMédia de renda para {var} ao longo dos anos:")
    print(media_renda)
````
