Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Resolução


# Número de acertos dos alunos (exemplo aleatório)
np.random.seed(42)  # Para reprodutibilidade
dados = np.random.randint(0, [24, 18, 8, 8, 8, 12, 12], size=(100, 7))

# Criando DataFrame com os acertos por assunto
df = pd.DataFrame(dados, columns=['Matemática', 'Português', 'Geografia', 'Inglês', 'História', 'Física', 'Química'])

# Mostrando os primeiros 5 alunos
df.head()

````

````python
# 2) Resolução

# Número de questões por assunto
questoes = pd.Series([24, 18, 8, 8, 8, 12, 12], index=df.columns)

# Calculando o percentual de acertos por assunto
percentual_por_assunto = df.div(questoes) * 100
percentual_por_assunto.head()

````

````python
# 3) Resolução

# Total de acertos por aluno
total_acertos = df.sum(axis=1)
total_acertos.head()
````

````python
# 4) Resolução
# Total de questões
total_questoes = questoes.sum()

# Percentual geral por aluno
percentual_geral = (total_acertos / total_questoes) * 100
percentual_geral.head()
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
1 # Resolução
# Alunos desqualificados (zeraram em Matemática, Física ou Química)
desqualificados = df[(df['Matemática'] == 0) | (df['Física'] == 0) | (df['Química'] == 0)]

# Mostrar o DataFrame de alunos desqualificados
desqualificados.head()
````

````python
2 # Resolução
# Contar o número de alunos desqualificados
num_desqualificados = len(desqualificados)
num_desqualificados
````

````python
3 # Resolução
# Média dos desqualificados em História e Geografia
media_hist_geo = desqualificados[['História', 'Geografia']].mean()
media_hist_geo
````

````python
4 # Resolução

# Alunos qualificados (não desqualificados)
qualificados = df[(df['Matemática'] > 0) & (df['Física'] > 0) & (df['Química'] > 0)]

# Total de acertos de cada aluno
total_acertos_qualificados = qualificados.sum(axis=1)

# Alunos que passaram para a segunda fase (nota maior que 45)
passaram_segunda_fase = qualificados[total_acertos_qualificados > 45]

# Mostrar o DataFrame dos alunos que passaram
passaram_segunda_fase.head()

````

````python
arquivo = 'https://opendatasus.saude.gov.br/dataset/covid-19-vacinacao/resource/5093679f-12c3-4d6b-b7bd-07694de54173/part-00000-105cf5c5-0cd3-4017-b5f0-acb8e6421676.c000.csv'
````

````python
import pandas as pd

# Lista de arquivos CSV
arquivos_csv = [
    'part-00000-105cf5c5-0cd3-4017-b5f0-acb8e6421676.c000.csv',
    'part-00001-105cf5c5-0cd3-4017-b5f0-acb8e6421676.c000.csv',
    'part-00002-105cf5c5-0cd3-4017-b5f0-acb8e6421676.c000.csv',
    'part-00003-105cf5c5-0cd3-4017-b5f0-acb8e6421676.c000.csv',
    'part-00004-105cf5c5-0cd3-4017-b5f0-acb8e6421676.c000.csv'
]

# Carregar e concatenar os CSVs em um único DataFrame usando o engine Python
df_list = [pd.read_csv(arquivo, sep=';', on_bad_lines='skip', engine='python') for arquivo in arquivos_csv]
df = pd.concat(df_list, ignore_index=True)

# Exibir as primeiras linhas para confirmar que os dados foram carregados corretamente
print(df.head())

# 1) Quantas vacinas estão registradas nessa base?
# Contar o número total de linhas no DataFrame, que corresponde ao número de vacinas registradas
total_vacinas = df.shape[0]

# Exibir o resultado
print(f"Total de vacinas registradas: {total_vacinas}")
````

````python
# 2) Quantos pacientes foram vacinados?

# Contar o número de vacinas registradas (número de linhas no DataFrame)
vacinas_registradas = len(df)

# Exibir o resultado
print(f"Total de vacinas registradas: {vacinas_registradas}")
````

````python
# 3) Quantos pacientes únicos tomaram a primeira dose?

# Contar o número de pacientes vacinados (valores únicos de paciente_id)
pacientes_vacinados = df['paciente_id'].nunique()

# Exibir o resultado
print(f"Total de pacientes vacinados: {pacientes_vacinados}")

````

````python
# 4) Quantos pacientes com menos de 18 anos foram vacinados?

# Remover espaços em branco dos nomes das colunas (caso existam)
df.columns = df.columns.str.strip()

# 4. Quantos pacientes com menos de 18 anos foram vacinados?
# Verifique se a coluna 'paciente_idade' existe
if 'paciente_idade' in df.columns:
    # Filtrar pacientes com menos de 18 anos
    pacientes_menores_18 = df[df['paciente_idade'] < 18]['paciente_id'].nunique()
    print(f"Total de pacientes com menos de 18 anos vacinados: {pacientes_menores_18}")
else:
    print("Coluna 'paciente_idade' não encontrada. Verifique o nome correto da coluna.")
````

````python
# 5) Quantos estabelecimentos aplicaram vacina no Acre?

# Remover espaços em branco dos nomes das colunas (caso existam)
df.columns = df.columns.str.strip()

# 5. Quantos estabelecimentos aplicaram vacina no Acre?
# Verifique se a coluna 'estabelecimento_noFantasia' existe
coluna_estabelecimento = 'estalecimento_noFantasia'  # Nome da coluna conforme fornecido
if coluna_estabelecimento in df.columns:
    # Contar valores únicos na coluna de estabelecimentos
    estabelecimentos = df[coluna_estabelecimento].nunique()
    print(f"Total de estabelecimentos que aplicaram vacina: {estabelecimentos}")
else:
    print(f"Coluna '{coluna_estabelecimento}' não encontrada. Verifique o nome correto da coluna.")

````

## 4) Vacinação II
Gere um *DataFrame* que contenha somente os estabelecimentos que aplicaram vcinas a menores de 18 anos. Nesse *DataFrame* devem conter somente os dados dos estabelecimentos, mais uma coluna sendo a quantidade de vacinas que o estabelecimento aplicou a menores de 18 anos.  
  
1. crie uma cópia do *DataFrame* original, contendo somente os registros de vacinas realizadas a menores de 18 anos.  
2. crie uma lista das colunas desse *DataFrame* com o atributo de *DataFrame* **.columns()**  
3. Nesse *DataFrame* faça uma contagem do campo ```vacina_categoria_nome```.
3. a partir da lista de colunas, escolha somente aquelas que são referentes ao estabelecimento, faça uma lista com esses valores.  
4. usando o método *.loc*, selecione somente essas variáveis  
5. Aplique o método **.drop_duplicates** e crie uma lista com uma linha para cada estabelecimento, com os dados do estabelecimento  

````python
# 1)

# Criar uma cópia do DataFrame original com registros de vacinas realizadas a menores de 18 anos
df_menores_18 = df[df['paciente_idade'] < 18].copy()

# Exibir as primeiras linhas para confirmar que os dados foram filtrados corretamente
print(df_menores_18.head())
````

````python
# 2)

# Criar uma lista das colunas desse DataFrame
colunas_df_menores_18 = df_menores_18.columns.tolist()

# Exibir a lista de colunas
print(colunas_df_menores_18)

````

````python
# 3)

# Contar o campo vacina_categoria_nome
contagem_vacina_categoria = df_menores_18['vacina_categoria_nome'].value_counts()

# Exibir a contagem
print(contagem_vacina_categoria)
````

````python
# 4)

# Lista de colunas referentes ao estabelecimento
colunas_estabelecimento = ['estabelecimento_razaoSocial', 'estalecimento_noFantasia',
                            'estabelecimento_municipio_nome', 'estabelecimento_uf']

# Selecionar somente essas variáveis usando .loc
df_estabelecimentos = df_menores_18.loc[:, colunas_estabelecimento]

# Exibir as primeiras linhas para confirmar a seleção
print(df_estabelecimentos.head())
````

````python
# 5)

# Contar a quantidade de vacinas aplicadas a menores de 18 anos por estabelecimento
df_contagem_vacinas = df_menores_18.groupby(colunas_estabelecimento).size().reset_index(name='quantidade_vacinas')

# Exibir o DataFrame final
print(df_contagem_vacinas)

````
